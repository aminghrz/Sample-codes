Sample dice by size n<br>
Get the mean<br>
Repeat 1000 times<br>
Conduct Shapiro-Wilk test to see if it follows Normal distribution<br>
Plot histogram of means<br>
Plot Normal distribution curve fitted to the data

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm , shapiro
dice = [1,2,3,4,5,6]
sample_size = [5,10,20,30,50,70,100,200,500,1000,2000]
exp_mean = []

fig, axs = plt.subplots(nrows=1,ncols=len(sample_size), figsize=(30,5))
plt.subplots_adjust(hspace=0.5)

for count, i in enumerate(sample_size):
    sample_mean = []
    for j in range(1000):
        sample_mean.append(np.mean(np.random.choice(dice,i)))
    stat , p_val = shapiro(sample_mean)
    mu, std = norm.fit(sample_mean)
    x = np.linspace(np.min(sample_mean),np.max(sample_mean), 100)
    p = norm.pdf(x, mu, std)
    axs[count].hist(sample_mean,bins=100)
    axs[count].plot(x, p, 'k')
    axs[count].set_title("SP test p-value:"+str(round(p_val,5)))
```
